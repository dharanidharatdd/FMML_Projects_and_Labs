#### Question 2 :
##### Given a list having Names, work_hours, and gender, Write a function to print name of the female worker that worked the most hours. Also how much do should she be paid if the pay is $ 20 per hour.

##### work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]

##### Answer : the female worker that worked the most hours is Penny and she should be paid 18400

```python
## 1. Question 1 Solution:
def rev_string(A:str,B:str):
  C=A[::-1] + B
  return(C)
rev_string('first','second')
```

```python
## 2. Question 2 Solution:
work_hours = [('Abby',100 , 'F'),('Billy',400, 'M'),('Cassie',800,'F'), ('Maggi',600,'F'),('Alex',500,'M'),('Raj',225,'M'),('Penny',920,'F'),('Ben',300,'M')]
def hardworker(work_hours):
  highest_earning_female = None
  max=0
  for name, hours, gender in work_hours:
    if gender == 'F' and hours > max:
      max = hours
      highest_earning_female = name

  if highest_earning_female:
    payment = max * 20
    print(f'The female worker that worked the most hours is {highest_earning_female} and she should be paid {payment}.')
  else:
    print('There are no female workers in the list.')
hardworker(work_hours)
```

#### Question 3 :

##### a)Create a 5x5 matrix of the following form,

##### [[1,1]
#####  [2,2]]

#####  i.e. each row is increasing and has repetive elements.

######  Hint : you can use hstack, vstack  etc.

##### b) find dot product of the matrix with any matrix. (Figure out the size/ shape of the matrix)

```python
import numpy as np

# Create a 2x2 matrix with the specified pattern
pattern_matrix = np.array([[1, 1], [2, 2]])

# Use np.tile to create a 5x5 matrix by repeating the pattern
matrix_5x5 = np.tile(pattern_matrix, (3, 3))

# Print the resulting 5x5 matrix
print("5x5 Matrix:")
print(matrix_5x5)
```
